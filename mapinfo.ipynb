```python
import csv
import pyads
import struct

PLC_IP = '5.74.170.110.1.1'
PLC_PORT = pyads.PORT_TC3PLC1
DATA_NAME = 'GVL.PositionTotalFloorForAGVC[5]'
ARRAY_SIZE = 10000

def read_data_from_csv(file_path):
    data = {
        'PositionNo': [],
        'PositionX': [],
        'PositionY': [],
        'PositionZ': [],
    }
    with open(file_path, mode='r', newline='') as file:
        reader = csv.DictReader(file)
        for row in reader:
            data['PositionNo'].append(int(row['Tag_ID']))
            data['PositionX'].append(int(row['PositionX']))
            data['PositionY'].append(int(row['PositionY']))
            data['PositionZ'].append(int(row['PositionZ']))
    return data

def write_data_to_plc(plc, data_name, data):
    for field, values in data.items():
        if len(values) > ARRAY_SIZE:
            raise ValueError(f"Data length for {field} ({len(values)}) exceeds the array size of the PLC ({ARRAY_SIZE}).")
        
        padded_data = values + [0] * (ARRAY_SIZE - len(values))
        
        if field == 'PositionNo':
            plc_type = pyads.PLCTYPE_WORD
        else:
            plc_type = pyads.PLCTYPE_DINT
        
        plc.write_by_name(f"{data_name}.{field}", padded_data, plc_type * ARRAY_SIZE)
        print(f"Data written successfully for {field}. First 10 values: {padded_data[:10]}")

def main():
    csv_file_path = 'C:\\Users\\noarj\\Downloads\\map_data (6).csv'
    data = read_data_from_csv(csv_file_path)
    
    print(f"Number of entries read from CSV: {len(data['PositionNo'])}")
    
    plc = pyads.Connection(PLC_IP, PLC_PORT)
    plc.open()
    
    try:
        write_data_to_plc(plc, DATA_NAME, data)
        print('All data written successfully.')
        
        read_data = plc.read_by_name(f"{DATA_NAME}.PositionNo", pyads.PLCTYPE_WORD * ARRAY_SIZE)
        print(f"First 10 PositionNo values read back: {read_data[:10]}")
        
    except Exception as e:
        print(f"An error occurred: {str(e)}")
        import traceback
        traceback.print_exc()
    
    finally:
        plc.close()

if __name__ == "__main__":
    main()
```
